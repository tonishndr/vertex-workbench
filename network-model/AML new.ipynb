## Load Bigquery dataset

```python
from google.cloud import bigquery
import pandas as pd
import networkx as nx
from networkx.algorithms import community

# Initialize a BigQuery client
client = bigquery.Client(project='poc-danamon-da')

# Define your query to retrieve transactions
query = """
SELECT *
FROM `poc-danamon-da.dataprep.dm_channel_trx_dob_new_grouped` limit 100
"""
```

```python
# Run the query and convert to a pandas DataFrame
query_job = client.query(query)
transactions_df = query_job.to_dataframe()
transactions_df.head()
```

```python
# Now, preprocess the DataFrame if necessary (e.g., filter, aggregate)
# ...

# Create a NetworkX graph from the DataFrame
G = nx.from_pandas_edgelist(transactions_df, source='sender', target='receiver', edge_attr=['amount'])

# Find triangles in the network
triangles = nx.triangles(G)

# Community detection using the Clauset-Newman-Moore greedy modularity maximization
communities = list(community.greedy_modularity_communities(G))

# You can now analyze the 'triangles' and 'communities' as needed
# ...

# For visualization, you can use matplotlib or other libraries as shown previously
# ...
```
